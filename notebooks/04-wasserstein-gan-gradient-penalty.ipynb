```python
import os, sys
sys.path.append(os.path.abspath(os.path.join('../src/')))
from w_gp_gan import *

# Enable Jupyter notebook plotting
%matplotlib inline

# Load in binzarized MNIST data, separate into data loaders
train_iter, val_iter, test_iter = get_data()

# Init model
model = WGANGP(image_size=784,
               hidden_dim=400,
               z_dim=20)
 
# Init trainer
trainer = WGANGPTrainer(model=model,
                        train_iter=train_iter,
                        val_iter=val_iter,
                        test_iter=test_iter,
                        viz=True)

# Train
trainer.train(num_epochs=25,
              G_lr=1e-4,
              D_lr=1e-4,
              D_steps=5)

# Show loss
trainer.viz_loss()
```
